**DROP VALUES THAT HAVE MORE THAN 40% MISSING VALUES**

```python
missing_values = data.isnull().sum()
missing_values_percent = 100 * data.isnull().sum() / len(data)
missing_values_table = pd.concat([missing_values, missing_values_percent], axis=1)
missing_values_table = missing_values_table.rename(columns={0: 'missing_values', 1: 'missing_values_percent'})
missing_values_table.sort_values(by='missing_values_percent', ascending=False).head(15)
```

```python
high_mising_values=missing_values_table.index[missing_values_table['missing_values_percent']>40]
list_missing_values=high_mising_values.tolist()
```

```python
df=data.drop(list_missing_values, axis=1)
df
```
